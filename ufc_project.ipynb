```python
#fucntion to create 'no of rounds'
def get_rounds(X):
    if X == 'No Time Limit':
        return 1
    
    else:
        rounds = int(X[0])
        return rounds
         
        #for eg: 3 Rnd (5-5-5)
        #order first split at '(' for rest of the string 

df['no_of_rounds'] = df['Format'].apply(get_rounds)
```

```python
#function to create total fight duration
def fight_duration(X):

    if X == 'No Time Limit':
        return 1
    
    elif '-' in X:
        rounds = int(X[0])
        round_duration = int(X.split('(')[1][0])
        if 'OT' not in X:
            duration = rounds*round_duration
        else:
            OT_duration = int(X.split('-')[1][0])
            if X.split('+')[1][1] == 'O':
                OT_rounds = 1
            else:
                OT_rounds = int(X.split('+')[1][1])
            
            duration  = rounds*round_duration + OT_rounds*OT_duration
    else:
        rounds = int(X[0])
        round_duration = int(X.split('(')[1][0:2])
        duration = rounds*round_duration
    
    return duration*60    
        #for eg: 3 Rnd (5-5-5)
        #order first split at '(' for rest of the string 

        
df['fight_duration'] = df['Format'].apply(fight_duration)
```

```python
def convert_to_cms(X):
    if X is np.NaN:
        return X
    elif len(X.split("'")) == 2:
        feet = float(X.split("'")[0])
        inches = int(X.split("'")[1].replace(' ', '').replace('"',''))
        return (feet * 30.48) + (inches * 2.54)
    else:
        #for reach conversion
        return float(X.replace('"','')) * 2.54
```

#### DEC: Decision (Dec) is a result of the fight or bout that does not end in a knockout in which the judges' scorecards are consulted to determine the winner; a majority of judges must agree on a result. A fight can either end in a win for an athlete, a draw, or a no decision.

#### SUB: also referred to as a "tap out" or "tapping out" - is often performed by visibly tapping the floor or the opponent with the hand or in some cases with the foot, to signal the opponent and/or the referee of the submission

#### KO/TKO: Knockout (KO) is when a fighter gets knocked out cold. (i.e.. From a standing to not standing position from receiving a strike.). Technical Knockout (TKO) is when a fighter is getting pummeled and is unable to defend him/herself further. The referee will step in and make a judgement call to end it and prevent the fighter from receiving any more unnecessary or permanent damage, and call it a TKO.

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from itertools import cycle
lw=1
```

```python
X1 = X_test
y1 = y_test
y1 = label_binarize(y1, classes=[0, 1, 2])
n_classes = y1.shape[1]
X1_train,X1_test,y1_train,y1_test = train_test_split(X1,y1,test_size = 0.20, random_state = 42)
pred_proba = best_xgb_classifier.predict_proba(X1_test)
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y1_test[:, i], pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
colors = cycle(['blue', 'red', 'green'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=lw,
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=lw)
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic for multi-class data')
plt.legend(loc="lower right")
plt.show()
```
